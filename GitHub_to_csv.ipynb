```python
import requests
import pandas as pd
import time
```

```python
#Git Hub search api base url
url = "https://api.github.com/search/issues"
```

```python
#find one page of issues from the GitHub API
def find_issues(page):
    response = requests.get(url, params={"q": f"repo:scipy/scipy is:issue","per_page": 100,"page": page},
                            headers={"Accept": "application/vnd.github.v3+json"})
    return response.json()
```

```python
# collect all scipy issues from Git Hub and save to csv
def get_all_issues():
    issues_list = []
    page = 1
    while True:
        data = find_issues(page)
        items = data.get("items", [])
        if not items:
            break
        for issue in items:
            issues_list.append({"source": "github_scipy","raw_text": issue.get("title", "").strip()})
        page += 1
        time.sleep(6)
    return issues_list

# run the issue collection and save to a csv
all_issues = get_all_issues()

df_github = pd.DataFrame(all_issues)
print(f"total issues collected: {len(df_github)}")

df_github.to_csv("/Users/davidmoriarty/Documents/github_issues_raw.csv", index=False)
print("saved to github_issues_raw.csv")
```
